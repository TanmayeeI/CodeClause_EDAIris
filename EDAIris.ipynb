```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
# Load dataset
from sklearn.datasets import load_iris

iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].apply(lambda x: iris.target_names[x])

data.head()
```

```python
print("Shape:", data.shape)
print("\nData Types:\n", data.dtypes)
print("\nMissing Values:\n", data.isnull().sum())
```

```python
data.describe()
```

```python
sns.countplot(x='species', data=data)
plt.title("Species Count")
plt.show()
```

```python
sns.pairplot(data, hue='species')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.drop('species', axis=1).corr(), annot=True, cmap='viridis')
plt.title("Correlation between features")
plt.show()
```

```python
for col in iris.feature_names:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='species', y=col, data=data)
    plt.title(f"{col} vs Species")
    plt.tight_layout()
    plt.show()
```

```python
for col in iris.feature_names:
    plt.figure(figsize=(6, 4))
    sns.violinplot(x='species', y=col, data=data)
    plt.title(f"{col} Distribution by Species")
    plt.tight_layout()
    plt.show()
```
